```python
#Your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

df = pd.read_csv('homepage_actions.csv')
df.head(50)
```

```python
df.action.value_counts()
```

```python
clicks = df.loc[df.action == 'click', 'id']
views = df.loc[df.action == 'view', 'id']
unique_clicks = set(clicks.unique())
unique_views = set(views.unique())
```

```python
unique_views
```

```python
click_no_view = len(unique_clicks - unique_views)
view_no_click = len(unique_views - unique_clicks)

print(click_no_view, view_no_click)
```

```python
control = df.loc[df.group == 'control', 'id']
experimental = df.loc[df.group == 'experiment', 'id']
controlgroup = set(control.unique())
expgroup = set(experimental.unique())

overlap = controlgroup.intersection(expgroup)
overlap
```

```python
print(len(controlgroup))
print(len(expgroup))
```

```python
df['counter'] = 0
df.head()
```

```python
df.loc[df.action == 'click', 'counter'] = 1
```

```python
df_control = df.loc[df.group == 'control']
df_exp = df.loc[df.group == 'experiment']

control_clicks = df_control.counter.sum()
exp_clicks = df_exp.counter.sum()

print(control_clicks, exp_clicks)
```

```python
control_rate = control_clicks/len(controlgroup)

exp_rate = exp_clicks/len(expgroup)

print(f'Control Rate: {control_rate}')
print(f'Experimental Rate: {exp_rate}')
```

```python
piv_control = df_control.pivot_table(index=['id'], columns=['action'], values='counter')
piv_control.fillna(0, inplace=True)
piv_control
```

```python
piv_exp = df_exp.pivot_table(index=['id'], columns=['action'], values='counter')
piv_exp.fillna(0, inplace=True)
piv_exp
```

## Conduct a Statistical Test

Conduct a statistical test to determine whether the experimental homepage was more effective than that of the control group.

```python
#Your code here
#Null Hypothesis: The average click rate on the control and experimental home pages is the same

#Alternative Hypothesis: The average click rate on the experimental home page is greater.

#Different size groups so we need to use a Welch's t test

import scipy.stats as stats
tstat, pval = stats.ttest_ind(piv_control.click, piv_exp.click, equal_var=False)
pval/2
```

## Verifying Results

One sensible formulation of the data to answer the hypothesis test above would be to create a binary variable representing each individual in the experiment and control group. This binary variable would represent whether or not that individual clicked on the homepage; 1 for they did and 0 if they did not. 

The variance for the number of successes in a sample of a binomial variable with n observations is given by:

## $n\bullet p (1-p)$

Given this, perform 3 steps to verify the results of your statistical test:
1. Calculate the expected number of clicks for the experiment group, if it had the same click-through rate as that of the control group. 
2. Calculate the number of standard deviations that the actual number of clicks was from this estimate. 
3. Finally, calculate a p-value using the normal distribution based on this z-score.

### Step 1:
Calculate the expected number of clicks for the experiment group, if it had the same click-through rate as that of the control group. 

```python
#Your code here
expected_clicks = (len(expgroup))*piv_control.click.mean()
expected_clicks
```

### Step 2:
Calculate the number of standard deviations that the actual number of clicks was from this estimate.

```python
#Your code here
n = len(expgroup)
p = control_rate
std = (n*p*(1-p))**0.5
std
```

### Step 3: 
Finally, calculate a p-value using the normal distribution based on this z-score.

```python
#Your code here
x = piv_exp.click.sum()
mu = expected_clicks
z = (x - mu)/std
z
```

```python
pval = 1 - stats.norm.cdf(z)
pval
```
